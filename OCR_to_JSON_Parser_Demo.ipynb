```python
# Install dependencies
!apt-get install -y tesseract-ocr
!pip install pytesseract pillow pymupdf

import pytesseract
from PIL import Image
import fitz  # PyMuPDF
import json

# OCR Function
def ocr_to_json(file_path):
    text_data = {}
    if file_path.lower().endswith(".pdf"):
        doc = fitz.open(file_path)
        for i, page in enumerate(doc):
            pix = page.get_pixmap()
            img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
            text = pytesseract.image_to_string(img)
            text_data[f"page_{i+1}"] = text.strip()
    else:
        image = Image.open(file_path)
        text = pytesseract.image_to_string(image)
        text_data["page_1"] = text.strip()
    return text_data

# Upload files
from google.colab import files
uploaded = files.upload()

for file_name in uploaded.keys():
    extracted_text = ocr_to_json(file_name)
    print(f"\nğŸ“„ Results for {file_name}:\n")
    print(json.dumps(extracted_text, indent=4, ensure_ascii=False))
```
